# Explorando gr√°ficos de barras circulares

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generando datos

# Valores positivos
df1 = pd.DataFrame(
        {'Name': ['item ' + str(i) for i in list(range(1, 51))],
         'Value': np.random.randint(low=0, high=4, size=50)
        })

array_val1 = np.array(df1["Value"])

#array_string1 = str(list(array_val1))

#print("array positivos " + array_string1)
print("array positivos")
print(array_val1)

# Valores negativos
df2 = pd.DataFrame(
        {'Name': ['item ' + str(i) for i in list(range(1, 51))],
         'Value': np.random.randint(low=-3, high=1, size=50)
        })

array_val2 = np.array(df2["Value"])

#array_string2 = str(list(array_val2))

#print("array negativos " + array_string2)
print("array negativos")
print(array_val2)


# Nombres
array_nombres = np.array(df2["Name"])

#array_string_nombres = str(list(array_nombres))

#print("Nombres " + array_string_nombres)
print("Nombres ")
print(array_nombres)
```

```python
# Configuraci√≥n del gr√°fico
angulos = np.linspace(0, 2 * np.pi, len(array_nombres), endpoint=False)
ancho = 2 * np.pi / len(array_nombres) # Gr√°fico se divide en la cantidad de "barras" que hay

# Crear la figura y los subplots
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Configurar los l√≠mites del eje radial
ax.set_ylim(-8, max(array_val1) + 1)

# Crear las barras exteriores
ax.bar(angulos, array_val1, width=ancho, alpha=1, color="#FFF300", edgecolor="black", linewidth=0.5)

# Crear las barras interiores
ax.bar(angulos, array_val2, width=ancho, alpha=0.7, color="gray", edgecolor="black", linewidth=0.5)

# Configurar las etiquetas de las categor√≠as
ax.set_xticks(angulos)
ax.set_xticklabels(array_nombres)

# A√±adir una grilla radial
ax.yaxis.grid(True, linestyle='-', linewidth=0.5, color='orange')

# Calcular el tama√±o del c√≠rculo a dibujar
#radio_circulo = max(array_string1)
radio_circulo = 0

# Dibujar el c√≠rculo conc√©ntrico alrededor de los dem√°s c√≠rculos
theta = np.linspace(0, 2 * np.pi, 100)
r = np.full_like(theta, radio_circulo)
#ax.fill(theta, r, color='black', alpha=0.2)
ax.plot(theta, r, color='black', linewidth=2)

# Eliminar el borde en l√≠nea negra
ax.spines['polar'].set_visible(False)


# remove grid
plt.axis('off')

# Mostrar el gr√°fico
plt.show()
```
