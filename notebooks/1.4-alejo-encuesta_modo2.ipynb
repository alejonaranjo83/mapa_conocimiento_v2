```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

# Define la variable 'index_persona' con el índice deseado
index_persona = 0

# Tu código para leer y procesar el DataFrame va aquí

# Define una función para obtener los ángulos a partir del número de valores
def get_angles(num_values):
    return np.linspace(0, 2 * np.pi, num_values, endpoint=False)

# Define la función get_label_rotation
def get_label_rotation(angle, offset):
    rotation = np.rad2deg(angle + offset)
    if angle <= np.pi:
        alignment = "right"
        rotation = rotation + 180
    else:
        alignment = "left"
    return rotation, alignment

# Define la función add_labels
def add_labels(angles, values, labels, offset, ax):
    padding = 0.1
    for angle, value, label in zip(angles, values, labels):
        angle = angle
        rotation, alignment = get_label_rotation(angle, offset)
        ax.text(
            x=angle,
            y=value + padding,
            s=label,
            ha=alignment,
            va="center",
            rotation=rotation,
            rotation_mode="anchor"
        )


# Inicializa la figura y los ejes
fig, ax = plt.subplots(figsize=(20, 10), subplot_kw={"projection": "polar"})

# Especificar offset
OFFSET = np.pi / 2

# Establece los límites para el eje radial (y)
ax.set_ylim(-8, 8)

# Ajusta el color de fondo de la figura y los ejes
ax.patch.set_facecolor('#ededed')
fig.patch.set_facecolor('#ededed')

# Remueve todos los ejes
ax.set_frame_on(False)

# Remueve la cuadrícula y las marcas de los ejes
ax.xaxis.grid(False)
ax.yaxis.grid(False)
ax.set_xticks([])
ax.set_yticks([])

# Obtiene los valores y etiquetas correspondientes a la persona seleccionada
posit_values = df_enc_proc_conv.loc[index_persona, "M1.1_posit":"M5.10_posit"].values
neg_values = df_enc_proc_conv.loc[index_persona, "M1.1_negat":"M5.10_negat"].values
labels = df_enc_proc_conv.columns[9:].values
num_values = len(posit_values)

# Obtiene los ángulos y el ancho de las barras
angles = get_angles(num_values)
width = 2 * np.pi / num_values

# Agrega las barras de valores positivos
ax.bar(
    angles,
    posit_values,
    width=width,
    color="#FFF300",
    # linewidth=2,
    # edgecolor="white"
)

# Agrega las barras de valores negativos
ax.bar(
    angles,
    neg_values,
    width=width,
    color='gray',
    # edgecolor="black",
    # linewidth=1
)

# Agregar círculo negro en el valor 0
circle_0 = patches.Circle((0, 0), radius=8, edgecolor='black', facecolor='none', lw=3, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_0)

# Agregando círculos grise de referencia
circle_8 = patches.Circle((0, 0), radius=9, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_8)

circle_9 = patches.Circle((0, 0), radius=10, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_9)

circle_10 = patches.Circle((0, 0), radius=11, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_10)

circle_3 = patches.Circle((0, 0), radius=7, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_3)

circle_2 = patches.Circle((0, 0), radius=6, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_2)

circle_1 = patches.Circle((0, 0), radius=5, edgecolor='gray', facecolor='none', lw=0.5, transform=ax.transData._b, zorder=10)
ax.add_patch(circle_1)

# Agregar texto en el centro del diagrama
center_text = df_enc_proc_conv.loc[index_persona, "Nombre"]
ax.text(0, -8, center_text, color='black', ha='center', va='center', fontsize=16)

# Agregar las etiquetas
add_labels(angles, posit_values, labels, OFFSET, ax)
add_labels(angles, neg_values, labels, OFFSET, ax)

plt.show()
```
