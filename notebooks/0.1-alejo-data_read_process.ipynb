## 1. Sacando porcentajes en función de 5 misiones

```python
# Creando columnas para sumar votos de misiones
col_m1 = fai.filter(regex=r'^M1').columns
suma_columnas_m1 = fai[col_m1].sum(axis=1)
suma_columnas_m1

col_m2 = fai.filter(regex=r'^M2').columns
suma_columnas_m2 = fai[col_m2].sum(axis=1)
suma_columnas_m2

col_m3 = fai.filter(regex=r'^M3').columns
suma_columnas_m3 = fai[col_m3].sum(axis=1)
suma_columnas_m3

col_m4 = fai.filter(regex=r'^M4').columns
suma_columnas_m4 = fai[col_m4].sum(axis=1)
suma_columnas_m4

col_m5 = fai.filter(regex=r'^M5').columns
suma_columnas_m5 = fai[col_m5].sum(axis=1)
suma_columnas_m5



# Llevando series independientes al Dataframe
fai = fai.join([suma_columnas_m1.rename('suma_m1_psna'), 
                suma_columnas_m2.rename('suma_m2_psna'), 
                suma_columnas_m3.rename('suma_m3_psna'), 
                suma_columnas_m4.rename('suma_m4_psna'), 
                suma_columnas_m5.rename('suma_m5_psna')])




# Sumar los valores de las columnas en cada fila
fai['suma_total_psna'] = fai[['suma_m1_psna', 
                              'suma_m2_psna', 
                              'suma_m3_psna', 
                              'suma_m4_psna', 
                              'suma_m5_psna']].sum(axis=1)

# Calcular el porcentaje de cada valor con relación a la suma total y redondear esa cantidad a una sola cifra
fai['porc_m1_psna'] = (fai['suma_m1_psna'] / fai['suma_total_psna'] * 100).round(1)
fai['porc_m2_psna'] = (fai['suma_m2_psna'] / fai['suma_total_psna'] * 100).round(1)
fai['porc_m3_psna'] = (fai['suma_m3_psna'] / fai['suma_total_psna'] * 100).round(1)
fai['porc_m4_psna'] = (fai['suma_m4_psna'] / fai['suma_total_psna'] * 100).round(1)
fai['porc_m5_psna'] = (fai['suma_m5_psna'] / fai['suma_total_psna'] * 100).round(1)


fai
```

## 2. Perfil general FAI

```python
# Creando DF resumen de FAI
df_fai_porcent = fai.groupby("Unidad")[['suma_m1_psna', 'suma_m2_psna', 'suma_m3_psna', 'suma_m4_psna', 'suma_m5_psna']].sum()

# Renombrando columnas
df_fai_porcent = df_fai_porcent.rename(columns={
    "suma_m1_psna": "suma_m1_fai",
    "suma_m2_psna": "suma_m2_fai",
    "suma_m3_psna": "suma_m3_fai",
    "suma_m4_psna": "suma_m4_fai",
    "suma_m5_psna": "suma_m5_fai"
})

# Sumar los valores de las columnas en cada fila
df_fai_porcent['suma_total_fai'] = df_fai_porcent[['suma_m1_fai', 
                              'suma_m2_fai', 
                              'suma_m3_fai', 
                              'suma_m4_fai', 
                              'suma_m5_fai']].sum(axis=1)

# Calcular el porcentaje de cada valor con relación a la suma total y redondear esa cantidad a una sola cifra
df_fai_porcent['porc_m1_fai'] = (df_fai_porcent['suma_m1_fai'] / df_fai_porcent['suma_total_fai'] * 100).round()
df_fai_porcent['porc_m2_fai'] = (df_fai_porcent['suma_m2_fai'] / df_fai_porcent['suma_total_fai'] * 100).round()
df_fai_porcent['porc_m3_fai'] = (df_fai_porcent['suma_m3_fai'] / df_fai_porcent['suma_total_fai'] * 100).round()
df_fai_porcent['porc_m4_fai'] = (df_fai_porcent['suma_m4_fai'] / df_fai_porcent['suma_total_fai'] * 100).round()
df_fai_porcent['porc_m5_fai'] = (df_fai_porcent['suma_m5_fai'] / df_fai_porcent['suma_total_fai'] * 100).round()

df_fai_porcent = df_fai_porcent.reset_index()


#df_fai_porcent


# Creando valores numéricos promedio por cada proyecto
proyectos_col_fai = [
    "M1.1", "M1.2", "M1.3", "M1.4", "M1.5", "M1.6", "M1.7", "M1.8", "M1.9", "M1.10",
    "M2.1", "M2.2", "M2.3", "M2.4", "M2.5", "M2.6", "M2.7", "M2.8", "M2.9", "M2.10",
    "M3.1", "M3.2", "M3.3", "M3.4", "M3.5", "M3.6", "M3.7", "M3.8", "M3.9", "M3.10",
    "M4.1", "M4.2", "M4.3", "M4.4", "M4.5", "M4.6", "M4.7", "M4.8", "M4.9", "M4.10",
    "M5.1", "M5.2", "M5.3", "M5.4", "M5.5", "M5.6", "M5.7", "M5.8", "M5.9", "M5.10",
]

promedio_proyectos_fai = fai[proyectos_col_fai].mean().apply(lambda x: math.ceil(x * 2) / 2).astype(int)

promedio_proyectos_fai = pd.DataFrame(promedio_proyectos_fai).transpose()
promedio_proyectos_fai.columns = proyectos_col_fai

promedio_proyectos_fai


# Uniendo ambos DFs

df_fai_perfil = pd.concat([promedio_proyectos_fai, df_fai_porcent], axis=1)

df_fai_perfil = df_fai_perfil.set_index("Unidad")

df_fai_perfil

```
